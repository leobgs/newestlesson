```python
#operator aritmatika operator yang digunakan untuk melakukan operasi matematika 

#variable
bil1= 12
bil= 5

#operator aritmatika penjumlahan(+)
penjumlahan= bil1 + bil
print("hasil dari", bil1, "+", bil, "adalah=", penjumlahan)

#operator aritmatika pengurangan(-)
pengurangan= bil1 - bil
print("hasil dari", bil1, "-", bil, "adalah=", pengurangan)

#operator aritmatika perkalian (x/*)
perkalian= bil1 * bil
print("hasil dari", bil1, "*", bil, "adalah=", perkalian)

#operator aritmatika pembagian (/)
pembagian= bil1 / bil
print("hasil dari", bil1, "/", bil, "adalah=", pembagian)

#operator aritmatika pemangkatan (**)
pemangkatan= bil1 ** bil
print("hasil dari", bil1, "**", bil, "adalah=", pemangkatan)

#operator aritmatika pembagian bulat/hasilnya bilangan bulat (//)
bagibulat= bil1 // bil
print("hasil dari", bil1, "//", bil, "adalah=", bagibulat)

#operator aritmatika modulus menampilkan sisa hasil bagi(%)
modul= bil1 % bil 
print("hasil dari", bil1, "%", bil, "adalah=", modul)
```

```python
#operator perbandingan adalah operator yang membandingkan dua buah nilai
#hasilnya True atau False tergantung kondisi

#memasukan variabel dengan fungsi input()
nilai1= int(input("Masukan nilai pertama: "))
nilai2= int(input("masukan Nilai kedua: "))

#operator perbandingan lebih dari (>)
optr_lebih= nilai2 > nilai1
print(nilai2, ">", nilai1, "adalah:", optr_lebih)

#operator perbandingan kurang dari (<)
optr_krg= nilai2 < nilai1
print(nilai2, "<", nilai1, "adalah:", optr_krg)

#operator perbandingan sama dengan (==)
optr_sama= nilai1 == nilai2
print(nilai1, "==", nilai2, "adalah:", optr_sama)

#operator lebih atau sama dengan (>=)
optr_lebihsama= nilai2 >= nilai1
print(nilai2, ">=", nilai1, "adalah:", optr_lebihsama)

#operator kurang atau sama dengan (<=)
optr_krgsama= nilai2 <= nilai1
print(nilai2, "<=", nilai1, "adalah:", optr_krgsama)

#operator tidak sama dengan (!=)
optr_samadengan= nilai2 != nilai1
print(nilai2, "!=", nilai1, "adalah:", optr_samadengan)
```
